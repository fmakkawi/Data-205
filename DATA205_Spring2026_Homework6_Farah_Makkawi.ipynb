```python
import pandas as pd
import sqlite3
```

```python
athlete_events = pd.read_csv('athlete_events (1).csv')
host_city_locations = pd.read_csv('host_city_locations.csv')

athlete_events.head()
```

```python
conn = sqlite3.connect('olympics.db')

athlete_events.to_sql('athlete_events', conn, if_exists='replace', index=False)
host_city_locations.to_sql('host_city_locations', conn, if_exists='replace', index=False)

print('Database created successfully.')
```

Question 1: How many records are in athlete_events?

```python
query1 = '''
SELECT COUNT(*) AS total_records
FROM athlete_events;
'''

pd.read_sql_query(query1, conn)
```

There are 271,116 records in the athlete_events table.

Question 2: How many different athletes competed in Olympics held in London?

```python
query2 = '''
SELECT COUNT(DISTINCT ID) AS london_athletes
FROM athlete_events
WHERE City = 'London';
'''

pd.read_sql_query(query2, conn)
```

16,941 different athletes competed in Olympics held in London.

Question 3: Which Olympics host city is furthest north?

```python
query3 = '''
SELECT City, lat
FROM host_city_locations
WHERE CAST(lat AS REAL) = (
    SELECT MAX(CAST(lat AS REAL))
    FROM host_city_locations
);
'''

pd.read_sql_query(query3, conn)
```

Lillehammer is the Olympics host city that is furthest north.

Question 4: What is the list of host cities with no duplicates?

```python
query4 = '''
SELECT DISTINCT City
FROM host_city_locations
ORDER BY City;
'''

host_cities = pd.read_sql_query(query4, conn)
host_cities
```

```python
host_cities.to_csv('host_cities_FarahMakkawi.csv', index=False)
print('host_cities_FarahMakkawi.csv saved successfully.')
```
